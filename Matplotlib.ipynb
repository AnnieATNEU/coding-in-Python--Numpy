```python
company=['GOOGL','AMZN','MSFT', 'FB']
revenue=[90,136,89,27]
profit=[40, 2, 34, 12]

xpos = np.arange(len(company))
xpos
```

```python
plt.bar(ypos,revenue)
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, label="Revenue")
plt.bar(xpos+0.2, profit,label=" Profit")
plt.legend()
```

# Rearranging side by side bar graph and setting its width 

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, label="Revenue")
plt.bar(xpos+0.2, profit,label=" Profit")
plt.legend()
```

```python
plt.xticks(ypos,company)
plt.ylabel("revenue(bin)")
plt.title("US Tech Stocks")
plt.bar(xpos-0.2,revenue, width= 0.4, label="Revenue")
plt.bar(xpos+0.2, profit,width= 0.4,label="Profit")
plt.legend()
```

```python
plt.yticks(ypos,company)
plt.title("US Tech Stocks")
plt.barh(xpos-0.2,revenue,label="Revenue")
plt.barh(xpos+0.2, profit,label="Profit")
plt.legend()
```

```python
import numpy as np
import matplotlib.pyplot as plt

rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5), marker,
             label="marker='{0}'".format(marker))
plt.legend(numpoints=1)
plt.xlim(0, 1.8);
```

```python
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, '-ok');
```

```python
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar();  # show color scale
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
features = iris.data.T

plt.scatter(features[0], features[1], alpha=0.2,
            s=100*features[3], c=iris.target, cmap='viridis')
plt.xlabel(iris.feature_names[0])
plt.ylabel(iris.feature_names[1]);
```
