```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0] #Pull the "Apples" wedge 0.2 from the center of the pie:

plt.pie(y, labels = mylabels, explode = myexplode)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]  #Pull the "Apples" wedge 0.2 from the center of the pie:

plt.pie(y, labels = mylabels, explode = myexplode, shadow = True) #add a shadow
plt.show() 
```

```python
#changing color
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["#FF0000", "purple", "yellow", "#9CFF00"]

myexplode = [0, 0, 0.2, 0]  #Pull the "Cherries" wedge 0.2 from the center of the pie:
plt.pie(y, labels = mylabels, explode = myexplode, colors = mycolors,  shadow = True)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["#FF0000", "purple", "yellow", "#9CFF00"]
myexplode = [0, 0, 0, 0.2] #Pull the "dates" wedge 0.2 from the center of the pie:
plt.pie(y, labels = mylabels,explode = myexplode, colors = mycolors, shadow = True)
plt.show() 
```

```python
#legend Legend
#To add a list of explanation for each wedge, use the legend() function:

mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
mycolors = ["#FF0000", "purple", "yellow", "#9CFF00"]
myexplode = [0, 0, 0, 0.2] #Pull the "dates" wedge 0.2 from the center of the pie:
plt.pie(y, labels = mylabels,explode = myexplode, colors = mycolors, shadow = True)

plt.legend() #here is where we added it
plt.show() 
```
